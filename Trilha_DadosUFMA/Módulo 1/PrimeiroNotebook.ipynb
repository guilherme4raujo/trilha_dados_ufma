```python
a = float(input('entre com o lado A:'))
b = float(input('entre com o lado B:'))
c = float(input('entre com o lado C:'))

def triangulo_valido(a,b,c):
  resultado =  a + b > c and b + c > a and a + c > b
  return resultado

if triangulo_valido(a,b,c):
  if (a == b == c):
    print('Este triangulo válido é EQUILATERO')
  elif ((a == b or b == c or c == b) and not (a == b == c)):
    print('Esse triângulo válido é ISÓRCELES')
  else:
    print('Esse triangulo válido é ESCALENO')
else:
  print('Esse triangulo não é válido')



```
